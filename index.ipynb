```python
import yfinance as yf
import matplotlib.pyplot as plt
import datetime as dt
import pandas as pd
```

```python
start = dt.datetime(2015, 1, 1)
end = dt.datetime(2024, 9, 1)

# df = yf.download('SANB11.SA', start=start, end=end)
```

```python
# Lista de ações com seus respectivos setores

acoes = [
    {"setor": "Tecnologia", "nome": "Totvs S.A.", "ticker": "TOTS3.SA"},
    {"setor": "Tecnologia","nome": "Positivo Tecnologia S.A.","ticker": "POSI3.SA"},
    {"setor": "Financeiro", "nome": "Banco do Brasil S.A.", "ticker": "BBAS3.SA"},
    {"setor": "Financeiro", "nome": "Itaú Unibanco Holding S.A.", "ticker": "ITUB4.SA"},
    {"setor": "Energia", "nome": "Petrobras (Petroleo Brasileiro S.A.)", "ticker": "PETR3.SA"},
    {"setor": "Energia", "nome": "Eletrobras (Centrais Elétricas Brasileiras S.A.)", "ticker": "ELET3.SA"},
    {"setor": "Saúde", "nome": "Raia Drogasil S.A.", "ticker": "RADL3.SA"},
    {"setor": "Saúde", "nome": "Fleury S.A.", "ticker": "FLRY3.SA"},
    {"setor": "Consumo", "nome": "Ambev S.A.", "ticker": "ABEV3.SA"},
    {"setor": "Consumo", "nome": "Lojas Renner S.A.", "ticker": "LREN3.SA"},
    {"setor": "Materiais", "nome": "Vale S.A.", "ticker": "VALE3.SA"},
    {"setor": "Materiais", "nome": "Suzano S.A.", "ticker": "SUZB3.SA"},
    {"setor": "Imobiliário", "nome": "Cyrela Brazil Realty S.A.", "ticker": "CYRE3.SA"},
    {"setor": "Imobiliário", "nome": "MRV Engenharia S.A.", "ticker": "MRVE3.SA"},
    {"setor": "Industrial", "nome": "Weg S.A.", "ticker": "WEGE3.SA"},
    {"setor": "Industrial", "nome": "Embraer S.A.", "ticker": "EMBR3.SA"},
]
```

```python
# Criando um DataFrame vazio
df_final = pd.DataFrame()

# Baixando os dados para cada ação e concatenando no DataFrame
for acao in acoes:
    ticker = acao['ticker']
    # Baixando os dados históricos da ação
    dados = yf.download(ticker, start=start, end=end)
    # Selecionando apenas a coluna 'Close' e renomeando com o ticker da ação
    df_acao_downloaded = dados[['Close']].rename(columns={'Close': ticker})
    # Concatenando os dados com o DataFrame final
    df_final = pd.concat([df_final, df_acao_downloaded], axis=1)
```
